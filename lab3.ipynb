```python
Step 1: Translation
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("car.jpg")
h, w = image.shape[:2]

tx, ty = 200, 120
M = np.float32([[1, 0, tx],
                [0, 1, ty]])

translated_image = cv2.warpAffine(
    image, M, (w, h),
    borderMode=cv2.BORDER_CONSTANT,
    borderValue=(255, 255, 255)
)

plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title(f"Translated Image (tx={tx}, ty={ty})")
plt.axis("off")
plt.show()
```

```python
Step 2: Rotation
```

```python
# 90 degrees clockwise
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title("Rotated Image (90 degrees)")
plt.axis("off")
plt.show()
print("Rotated:", rotated_image.shape)
```

```python
Step 3: Scaling
```

```python
Factor = 2
scaled_image = cv2.resize(image, None, fx=Factor, fy=Factor, interpolation=cv2.INTER_LINEAR)

plt.imshow(cv2.cvtColor(scaled_image, cv2.COLOR_BGR2RGB))
plt.title("Scaled Image (2x)")
plt.axis("off")
plt.show()

print("Original:", image.shape)
print("Scaled:", scaled_image.shape)
```

```python
Step 4: Shearing
```

```python
m = 0.67
h, w = image.shape[:2]

M = np.float32([[1, m, 0],
                [0, 1, 0]])

sheared_image = cv2.warpAffine(
    image, M, (w, h),
    borderMode=cv2.BORDER_REFLECT
)

plt.imshow(cv2.cvtColor(sheared_image, cv2.COLOR_BGR2RGB))
plt.title(f"Sheared Image (x-axis, factor {m})")
plt.axis("off")
plt.show()
```
