```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('car.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
subtracted_image = cv2.subtract(image_RGB, 150)
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(subtracted_image)
plt.title('Subtracted Image (-150)')
plt.axis('off')
plt.show()
```

```python
import cv2
image_color = cv2.imread('car.jpg')
image_add = cv2.add(image_color, 100)
cv2.imshow('added', image_add)
cv2.waitKey(-1)
```

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('car.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
added_image = cv2.add(image_RGB, 100)
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(added_image)
plt.title('Added Image (100)')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('car.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
multiplied_image = cv2.multiply(image_RGB, 2)
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(multiplied_image)
plt.title('Multiplied Image 2')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('car.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
multiplied_image = cv2.divide(image_RGB, 2)
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(multiplied_image)
plt.title('Divided Image 2')
plt.axis('off')
plt.show()
```
